```python
#    print     .
from defusedxml import DTDForbidden


def is_position_safe(N, M, position):
    # N * N    
    # 1.        
    # 2.    
    
    # position == (0, 0)   
    if position == (0, 0):
        if M == 0 or M == 2:
            return False

    # position == (0, N)  
    if position == (0, N):
        if M == 0 or M == 3:
            return False
    
    # position == (N, 0)   
    if position == (N, 0):
        if M == 2 or M == 1:
            return False

    # position == (N, N)           
    if position == (N, N):
        if M == 1 or M == 3:
            return False
    

    # position == (1 ~ N-1, N)   
    for i in range(1, N):
        if position == (i, N):
            if M == 3:
                return False


    # position == (0, 1 ~ N-1)   
    for i in range(1, N):
        if position == (0, i):
            if M == 0:
                return False

    # position == (1 ~ N-1, 0)   
    for i in range(1, N):
        if position == (i, 0):
            if M == 2:
                return False

    # position == (N, 1 ~ N-1)   
    for i in range(1, N):
        if position == (N, i):
            if M == 1:
                return False
    
    return True 
    
        


#      .
if __name__ == '__main__':
    print(is_position_safe(3, 1, (0, 0))) # True
    print(is_position_safe(3, 0, (0, 0))) # False
    
```

```python
#     .




#pattern = [list(map(str, input())) for _ in range(2)]

def is_pattern(m):
    # pattern 
    pattern = []
    rlt = ''
    for row in range(2):
        row = list(map(str, input())) 
        pattern.append(row)

    # a = m[0][2:4]
    # print(a)    
    # m  pattern   - 1   (1),  
    cnt = 0
    for i in m: #  
        for j in range(len(m[0])-1):
            if i[j:j+2] == pattern[0]:
                #print(f': {pattern[0]}')
                cnt += 1
            if i[j:j+2] == pattern[1]:
                cnt += 1

    if cnt > 0:
        rlt = f"({cnt})"
    else:
        rlt = ""

    return rlt


map_1 = [['A', 'B', 'G', 'K'], ['T', 'T', 'A', 'B'], ['A', 'C', 'C', 'D']]
ans = is_pattern(map_1)
print(f'"{ans}"')
```
